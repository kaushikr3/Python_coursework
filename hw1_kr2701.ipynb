## 2. Fahrenheit to Celsius conversion
A. Convert the temperature 23 degrees in Fahrenheit to Celsius.

```python
#function that converts F to C
def f2c(f):
  assert isinstance(f,int), "input must be an integer"
  print("The temperature in Farenheit is", f)
  c = ((f-32)*5)/9
  print("The temperature in Celsius is", c)
  
  return(c)
```

```python
f2c(23)
```
